```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col = ['area', 'perimeter', 'compactness', 'length', 'width', 'asymetry', 'groove', 'class']
df = pd.read_csv('/content/seeds_dataset.txt',names = col, sep='\s+')
```

```python
df.head()
```

```python
for i in range(len(col)-1):
  for j in range(i+1, len(col)-1):
    x_label = col[i]
    y_label = col[j]
    sns.scatterplot(x=x_label, y=y_label, data=df, hue='class')
    plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
x = 'compactness'
y = 'asymetry'
X = df[[x,y]].values
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
clusters = kmeans.labels_
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1,1))), columns=[x,y,'class'])
```

```python
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df)
plt.plot()
```

```python
sns.scatterplot(x=x, y=y, hue='class', data=df)
plt.show()
```

#Higher Dimensions

```python
X = df[col[:-1]].values
```

```python
kmeans = KMeans(n_clusters = 3).fit(X)
```

```python
cluster_df = pd.DataFrame(np.hstack((X, kmeans.labels_.reshape(-1,1))), columns=df.columns)
```

```python
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df)
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
transformed_x = pca.fit_transform(X)
```

```python
plt.scatter(transformed_x[:,0], transformed_x[:,1])
plt.show()
```

```python
kmeans_pca = pd.DataFrame(np.hstack((transformed_x, kmeans.labels_.reshape(-1,1))), columns=['pca1', 'pca2', 'class'])
```

```python
truth_pca_df = pd.DataFrame(np.hstack((transformed_x, df['class'].values.reshape(-1,1))), columns=['pca1', 'pca2', 'class'])
```

```python
sns.scatterplot(x='pca1', y='pca2', hue='class', data=kmeans_pca)
```

```python
sns.scatterplot(x='pca1', y='pca2', hue='class', data=truth_pca_df)
```
